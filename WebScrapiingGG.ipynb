```python

columns = ["match", "team", "battingposition", "batsmanname", "dismissal", "runs", "balls", "fours", "sixes", "strikerate"]
data = []

def clean_text(text):
    text = re.sub(r'[^\x00-\x7F]+', '', text) 
    text = re.sub(r'\(.*?\)', '', text)  
    return text.strip()

# Iterate through the match IDs
for i in range(1415701, 1415756):
    API_KEY = "f9479ced4d7be2f52fc2fe30a1206f67"
    url = f"http://api.scraperapi.com?api_key={API_KEY}&url=https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/united-states-of-america-vs-canada-1st-match-group-a-{i}/full-scorecard"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "lxml")
        
        team_names = []
        innings = soup.find_all("div", class_="ds-rounded-lg ds-mt-2")
        
        for inning in innings:
            team_name_elem = inning.find("span", class_="ds-text-title-xs ds-font-bold ds-capitalize")
            if team_name_elem:
                team_names.append(team_name_elem.text.strip())
        
        # Check if we found valid team names
        if len(team_names) < 2:
            print(f"ABANDONED MATCH ID {i}")
            continue  
        
        match_title = f"{team_names[0]} vs {team_names[1]}"
        for inning in innings:
            team_name_elem = inning.find("span", class_="ds-text-title-xs ds-font-bold ds-capitalize")
            if team_name_elem:
                team_name = team_name_elem.text.strip()
                batting_table = inning.find("table", class_="ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table")
                
                if batting_table:
                    rows = batting_table.find_all("tr")
                    batting_position = 1
                    for row in rows:
                        cols = row.find_all("td")
                        if len(cols) >= 7:
                            batsman_name = clean_text(cols[0].text.strip())
                            dismissal = cols[1].text.strip()
                            R = cols[2].text.strip()
                            B = cols[3].text.strip()
                            fours = cols[5].text.strip()
                            sixes = cols[6].text.strip()
                            SR = cols[7].text.strip()
                            
                            data.append([match_title, team_name, batting_position, batsman_name, dismissal, R, B, fours, sixes, SR])
                            batting_position += 1


df_batting = pd.DataFrame(data, columns=columns)


df_batting
```

```python
columns = ["match", "team", "bowlerName", "over", "maiden", "runs", "wickets", "economy", "zeros", "fours", "sixes", "wides", "noballs"]
data = []

for i in range(1415701, 1415756):
    API_KEY = "f9479ced4d7be2f52fc2fe30a1206f67"
    url = f"http://api.scraperapi.com?api_key={API_KEY}&url=https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/united-states-of-america-vs-canada-1st-match-group-a-{i}/full-scorecard"

    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)

    if response.status_code == 200: 
        soup = BeautifulSoup(response.content, "lxml")

        team_names = []
        innings = soup.find_all("div", class_="ds-rounded-lg ds-mt-2")

        for inning in innings:
            team_name_elem = inning.find("span", class_="ds-text-title-xs ds-font-bold ds-capitalize")
            if team_name_elem:
                team_names.append(team_name_elem.text.strip())

        if len(team_names) < 2:
            print(f"ABANDONED MATCH ID {i}")
            continue  

        match_title = f"{team_names[0]} vs {team_names[1]}"

        for idx, inning in enumerate(innings):
            batting_team_elem = inning.find("span", class_="ds-text-title-xs ds-font-bold ds-capitalize")
            if batting_team_elem:
                batting_team = batting_team_elem.text.strip()

                bowling_team = team_names[1] if batting_team == team_names[0] else team_names[0]

            bowling_table = inning.find("table", class_="ds-w-full ds-table ds-table-md ds-table-auto")

            if bowling_table:
                rows = bowling_table.find_all("tr")
                for row in rows:
                    cols = row.find_all("td")
                    if len(cols) >= 11:
                        bowling = cols[0].text.strip()
                        O = cols[1].text.strip()
                        M = cols[2].text.strip()
                        R = cols[3].text.strip()
                        W = cols[4].text.strip()
                        economy = cols[5].text.strip()
                        zeros = cols[6].text.strip()
                        fours = cols[7].text.strip()
                        sixes = cols[8].text.strip()
                        WD = cols[9].text.strip()
                        NB = cols[10].text.strip()
                        
                        
                        data.append([match_title, bowling_team, bowling, O, M, R, W, economy, zeros, fours, sixes, WD, NB])


df_bowling = pd.DataFrame(data, columns=columns)


df_bowling
```

```python
columns = ["Playername", "Team", "Image", "Battingstyle", "Bowlingstyle", "Playingrole"]
player_data = []
unique_players = set()
player_profiles = {} 
team_dict = {} 



def clean_text(text):
    text = re.sub(r'[^\x00-\x7F]+', '', text)  
    text = re.sub(r'\(.*?\)', '', text)  
    return text.strip()

for i in range(1415701, 1415756):
    API_KEY = "f9479ced4d7be2f52fc2fe30a1206f67"
    url = f"http://api.scraperapi.com?api_key={API_KEY}&url=https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/united-states-of-america-vs-canada-1st-match-group-a-{i}/full-scorecard"
    headers= {"User-Agent": "Chrome/120.0.0.0 Safari/537.36"}
    response = requests.get(url,headers)

    if response.status_code == 200:
        response.encoding = 'utf-8'
        soup = BeautifulSoup(response.content, "html.parser")

       
        innings = soup.find_all("div", class_='ds-rounded-lg ds-mt-2')
        
        
        for index, inning in enumerate(innings):
            team_name = inning.find("span", class_="ds-text-title-xs ds-font-bold ds-capitalize")
            if team_name:
                team_name = team_name.text.strip()
                team_dict[index] = team_name

        
        for index, inning in enumerate(innings):
            team_name = team_dict.get(index, "Unknown Team")  

            
            batting_table = inning.find("table", class_="ds-w-full ds-table ds-table-md ds-table-auto ci-scorecard-table")
            if batting_table:
                for row in batting_table.find_all("tr"):
                    cols = row.find_all("td")
                    if len(cols) >= 7:
                        player_name = clean_text(cols[0].text.strip())

                        if player_name in unique_players:
                            continue 
                        unique_players.add(player_name)

                       
                        player_link_tag = cols[0].find("a")
                        if player_link_tag:
                            player_profile_url = "https://www.espncricinfo.com" + player_link_tag["href"]
                            og_url=player_profile_url = f"http://api.scraperapi.com?api_key={API_KEY}&url={player_profile_url}"


                            
                            if player_name not in player_profiles:
                                player_response = requests.get(og_url)
                                if player_response.status_code == 200:
                                    player_response.encoding = 'utf-8'
                                    player_soup = BeautifulSoup(player_response.content, "html.parser")

                                  
                                    image_tag = player_soup.find("img", alt=lambda x: x and player_name in x)
                                    image_url = image_tag['src'] if image_tag else "N/A"

                                    batting_style_p = player_soup.find("p", string="Batting Style")
                                    batting_style = batting_style_p.find_next("p").text.strip() if batting_style_p else "N/A"

                                    bowling_style_p = player_soup.find("p", string="Bowling Style")
                                    bowling_style = bowling_style_p.find_next("p").text.strip() if bowling_style_p else "N/A"

                                    playing_role_p = player_soup.find("p", string="Playing Role")
                                    playing_role = playing_role_p.find_next("p").text.strip() if playing_role_p else "N/A"

                                    player_profiles[player_name] = [player_name, team_name, image_url, batting_style, bowling_style, playing_role]

                            player_data.append(player_profiles[player_name])

       
        for index, inning in enumerate(innings):
            bowling_team = team_dict.get(1 - index, "Unknown Team")  
            if bowling_team == "Unknown Team":
                print(f"Warning: team_dict does not contain key {1 - index}, check team assignment.")

            bowling_table = inning.find("table", class_="ds-w-full ds-table ds-table-md ds-table-auto")
            if bowling_table:
                for row in bowling_table.find_all("tr"):
                    cols = row.find_all("td")
                    if len(cols) >= 2:
                        player_name = clean_text(cols[0].text.strip())

                        if player_name in unique_players:
                            continue  
                        unique_players.add(player_name)

                        player_link_tag = cols[0].find("a")
                        if player_link_tag:
                            player_profile_url = "https://www.espncricinfo.com" + player_link_tag["href"]
                            og_url=player_profile_url = f"http://api.scraperapi.com?api_key={API_KEY}&url={player_profile_url}"

                            if player_name not in player_profiles:
                                player_response = requests.get(og_url)
                                if player_response.status_code == 200:
                                    player_response.encoding = 'utf-8'
                                    player_soup = BeautifulSoup(player_response.content, "html.parser")

                                    image_tag = player_soup.find("img", alt=lambda x: x and player_name in x)
                                    image_url = image_tag['src'] if image_tag else "N/A"

                                    batting_style_p = player_soup.find("p", string="Batting Style")
                                    batting_style = batting_style_p.find_next("p").text.strip() if batting_style_p else "N/A"

                                    bowling_style_p = player_soup.find("p", string="Bowling Style")
                                    bowling_style = bowling_style_p.find_next("p").text.strip() if bowling_style_p else "N/A"

                                    playing_role_p = player_soup.find("p", string="Playing Role")
                                    playing_role = playing_role_p.find_next("p").text.strip() if playing_role_p else "N/A"

                                    player_profiles[player_name] = [player_name, bowling_team, image_url, batting_style, bowling_style, playing_role]

                            player_data.append(player_profiles[player_name])

df_players1 = pd.DataFrame(player_data, columns=columns)


df_players1
```
